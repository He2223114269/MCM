```python
import pandas as pd
import numpy as np
from statsmodels.tsa.api import VAR
from statsmodels.tools.eval_measures import rmse
from statsmodels.tsa.stattools import adfuller

# 1. 数据收集和准备

# 读取数据
data = pd.read_excel('数据.xlsx')
# data = np.log(data)
# 对于不平稳的变量进行一阶差分
data['农作物受灾面积'] = data['农作物受灾面积'].diff()
data['粮食产量'] = data['粮食产量'].diff()
data['平均温度'] = data['平均温度'].diff()
data['农业总产值'] = data['农业总产值'].diff()
data['平均降雨量'] = data['平均降雨量'].diff()
data['农业机械总动力'] = data['农业机械总动力'].diff()
data['农药使用量'] = data['农药使用量'].diff()
data['第一产业就业人员'] = data['第一产业就业人员'].diff()

data['粮食产量'] = data['粮食产量'].diff()
data['平均温度'] = data['平均温度'].diff()
data['农药使用量'] = data['农药使用量'].diff()
data['第一产业就业人员'] = data['第一产业就业人员'].diff()

data['粮食产量'] = data['粮食产量'].diff()
data['农药使用量'] = data['农药使用量'].diff()

# # 删除差分后产生的NaN值
data.dropna(inplace=True)

# 检查时间序列的平稳性
for column in data.columns:
    result = adfuller(data[column])
    print(f'ADF Statistic for {column}: {result[0]}, p-value: {result[1]}')

# 继续对其他变量进行差分操作
```

```python
import pandas as pd

# 创建一个 DataFrame 存储结果
adf_results = pd.DataFrame(columns=['ADF Statistic', 'p-value'])

for column in data.columns:
    result = adfuller(data[column])
    adf_results.loc[column] = [result[0], result[1]]

# print(adf_results)
adf_results.to_excel("adf_results.xlsx")
```

```python
# 筛选出平稳的变量
stable_variables = ['粮食产量', '农作物受灾面积', '平均温度', '平均降雨量', '农业机械总动力', '农药使用量', '第一产业就业人员']
data_stable = data[stable_variables]

# 3. 模型拟合
# 拟合VAR模型
model = VAR(data_stable)

# 选择合适的滞后阶数
lag_order = model.select_order()

# 获取最佳的滞后阶数
best_lag = lag_order.selected_orders['aic']

# 拟合模型
results = model.fit(best_lag)

# 4. 模型诊断
# 检查残差是否满足白噪声
print("Ljung-Box test for white noise in residuals:\n", results.test_whiteness())

# 5. 结果解释
# 打印模型系数
print("Model coefficients:\n", results.summary())


# 检查模型拟合是否成功
if results is None:
    print("Error: Model fitting failed!")
else:
    # 模型拟合成功，打印模型系数
    print("Model coefficients:\n", results.params) 

# 获取模型的预测值
predicted_values = results.fittedvalues

# 获取模型的观测值
observed_values = data_stable[best_lag:]

# 计算残差的均方根误差
print("RMSE of residuals:", rmse(observed_values, predicted_values))
```
