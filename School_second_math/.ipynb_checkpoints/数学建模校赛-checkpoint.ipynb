```python
import pandas as pd
import statsmodels.api as sm

# 加载数据
# data = log_data

# 提取自变量和因变量
X = data[[ '农作物受灾面积', '平均温度','平均降雨量', '第一产业就业人员','农业机械总动力', '农药使用量']]
y = data['粮食产量']

# 添加截距项
X = sm.add_constant(X)

# 创建线性回归模型
model = sm.OLS(y, X)

# 拟合模型
results = model.fit()

# 打印回归结果摘要
print(results.summary())
# results.summary()
```

OLS（普通最小二乘法）线性回归模型的结果摘要

1. **R-squared 和 Adj. R-squared**：R-squared是拟合模型所解释的因变量方差的比例，这个模型的R-squared为0.957，表明自变量能够解释因变量约95.7%的方差。Adjusted R-squared通过考虑模型中自变量的数量来调整R-squared，避免了过度拟合的问题。这个模型的Adjusted R-squared为0.920。

2. **F-statistic 和 Prob (F-statistic)**：F-statistic是模型的F统计量，用于检验模型中的自变量是否显著。这个模型的F-statistic为25.95，p-value为0.000191，表明模型整体显著。

3. **coef**：coef列给出了模型中每个自变量的系数估计值。例如，第三产业就业人员(万人)的系数估计值为0.6592，表示当第三产业就业人员增加一个单位时，粮食产量预期增加0.6592万吨。

4. **std err**：std err列给出了每个系数估计值的标准误差。

5. **t-statistic 和 P>|t|**：t-statistic是每个系数估计值除以其标准误差得到的t值，用于检验该系数是否显著不等于零。P>|t|是检验系数的双侧假设检验的p-value。例如，农作物受灾面积(千公顷)的P>|t|为0.091，表示其系数在10%的显著水平下不显著。

6. **Omnibus, Prob(Omnibus), Skew 和 Kurtosis**：这些统计量用于检验残差是否服从正态分布。Prob(Omnibus)表示Omnibus统计量的p-value，如果这个值很小，则表明残差不服从正态分布。

7. **Durbin-Watson**：Durbin-Watson统计量用于检验残差之间是否存在自相关性。它的取值范围是0到4，接近2表明不存在一阶自相关性。

8. **Jarque-Bera (JB)**：Jarque-Bera统计量也用于检验残差是否服从正态分布。

9. **Cond. No.**：条件数（Condition Number）用于衡量自变量之间的多重共线性程度。如果条件数较大，则表示存在较严重的多重共线性。
