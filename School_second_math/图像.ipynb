```python
from pyecharts.charts import Map, Tab
from pyecharts import options as opts
import pandas as pd

# te_ad = pd.read_excel("技术进步指数.xlsx",sheet_name = "综合评分")
import pandas as pd
te_ad = pd.read_excel("地图数据.xlsx")
te_ad

# 假设 te_ad 是一个 DataFrame 包含数据
# te_ad = pd.read_excel("地图数据.xlsx")

province = ['北京', '天津', '河北省', '山西省', '内蒙古自治区', '辽宁省', '吉林省', '黑龙江省', '上海', '江苏省',
       '浙江省', '安徽省', '福建省', '江西省', '山东省', '河南省', '湖北省', '湖南省', '广东省',
       '广西壮族自治区', '海南省', '重庆市', '四川省', '贵州省', '云南省', '西藏自治区', '陕西省', '甘肃省',
       '青海省', '宁夏回族自治区', '新疆维吾尔自治区']

# 创建一个 Tab 实例
tab = Tab()

# 遍历每个子列表，并为其创建一个地图实例，并将地图添加到 Tab 中
for i in range(8):
    data = [(province[j], te_ad.iloc[i][j]) for j in range(len(province))]
    map_ = (
        Map()
        .add("", data, maptype="china", zoom=1)
        .set_global_opts(
            title_opts=opts.TitleOpts(title=f"地图{i+1}"),
            visualmap_opts=opts.VisualMapOpts(max_=te_ad.iloc[i].max(), min_=te_ad.iloc[i].min())
        )
    )
    tab.add(map_, f"地图{i+1}")

# 显示 Tab
tab.render_notebook()
```
