```python
import pandas as pd
points_x = pd.read_excel("板凳龙螺旋为0.55米x坐标.xlsx")
points_x = points_x.drop(["Unnamed: 0"],axis = 1)
points_y = pd.read_excel("板凳龙螺旋为0.55米y坐标.xlsx")
points_y = points_y.drop(["Unnamed: 0"],axis = 1)
```

```python
points = pd.DataFrame()
points
```

```python
name_x = points_x.columns
name_y = points_y.columns
```

```python
name_x
```

```python
name_y
```

```python
for i in range(points_y.shape[1]):
    points[name_x[i]] = points_x[name_x[i]]
    points[name_y[i]] = points_y[name_y[i]]
```
