```python
import pandas as pd
points_x = pd.read_excel("板凳龙x盘出坐标.xlsx")
points_x = points_x.drop(["Unnamed: 0"],axis = 1)
points_y = pd.read_excel("板凳龙y盘出坐标.xlsx")
points_y = points_y.drop(["Unnamed: 0"],axis = 1)
```

```python
import matplotlib.pyplot as plt
# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体字体
plt.rcParams['axes.unicode_minus'] = False  # 解决坐标轴负号显示问题
a = 0  # 初始半径
p = 1.7  # 螺距
b = p / (2 * np.pi)  # b参数由螺距 p 确定
for i in range(10):
    # 提取数据
    pointsx = points_x.loc[i * 10].values
    pointsy = points_y.loc[i * 10].values

    # 创建一个新的图形和子图
    plt.figure(figsize=(8, 8))  # 调整图像尺寸为8x8
    
     # 生成螺线的角度范围
    theta = np.linspace(0, 2 * np.pi * 10, 1000)  # 从0到16圈

    # 计算螺线半径
    r = a + b * theta

    # 将极坐标转换为笛卡尔坐标
    x_spiral = r * np.cos(theta)
    y_spiral = r * np.sin(theta)
    # 绘制螺线
    plt.figure(figsize=(8, 8))
    plt.plot(x_spiral, y_spiral, label="盘入螺旋轨迹")
    plt.plot(-x_spiral, -y_spiral, label="盘出螺旋轨迹")
    plt.xlim([min(pointsx) - 1, max(pointsx) + 1])
    plt.ylim([min(pointsy) - 1, max(pointsy) + 1])

    # 绘制散点图
    plt.scatter(pointsx, pointsy, color='blue', label='板凳龙把手')

    # 绘制曲线图
    plt.plot(pointsx, pointsy, color='red', label='板凳龙')

    # 调整坐标轴范围（自动适应）
    plt.axis('equal')  # 确保x和y轴的比例一致
    plt.xlim([-20, 20])  # 手动设置x轴范围
    plt.ylim([-20, 20])  # 手动设置y轴范围

    # 添加标题和标签
    plt.title(f"{i}秒的板凳龙点迹")
    plt.xlabel("X values")
    plt.ylabel("Y values")

    # 显示图例
    plt.legend()

    # 显示图像
    plt.show()
```
