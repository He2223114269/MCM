```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams

# 设置中文字体
rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体字体
rcParams['axes.unicode_minus'] = False  # 解决坐标轴负号显示问题

# 参数设置
pitch = 55  # 螺距，单位：cm
b = pitch / (2 * np.pi)  # b 参数计算
a = 0  # 起始点可以设为 0
v_tangent = 1  # 切线速度，每秒 1 米
num_turns = 16  # 圈数

# 时间范围（假设每秒记录一次，直到螺线结束）
total_time = int(2 * np.pi * num_turns * b / v_tangent)  # 总时间 = 螺旋线长度 / 速度
time_steps = np.arange(0, total_time + 1, 1)  # 每秒一个时间点

# 初始化坐标数组
positions = []

# 计算每一时刻点的坐标
for t in time_steps:
    # 根据时间计算当前角度 theta
    theta_t = v_tangent * t / b
    
    # 极坐标方程计算 r
    r_t = a + b * theta_t
    
    # 将极坐标转换为笛卡尔坐标
    x_t = r_t * np.cos(theta_t)
    y_t = r_t * np.sin(theta_t)
    
    # 保存坐标 (x, y)
    positions.append([t, x_t, y_t])

# 转换为 DataFrame
df = pd.DataFrame(positions, columns=['时间 (秒)', 'X 位置 (米)', 'Y 位置 (米)'])

# 保存为 Excel 文件
excel_path = "data/spiral_positions.xlsx"
df.to_excel(excel_path, index=False)

# 绘制运动轨迹
x_vals = df['X 位置 (米)']
y_vals = df['Y 位置 (米)']
plt.figure(figsize=(6, 6))
plt.plot(x_vals, y_vals)
plt.title(f"沿螺旋线移动的点的位置（每秒 1 米的切线速度）")
plt.xlabel("X 位置 (米)")
plt.ylabel("Y 位置 (米)")

# 设置 x 和 y 轴的范围
plt.xlim([min(x_vals-100), max(x_vals+100)])
plt.ylim([min(y_vals-100), max(y_vals+100)])

plt.gca().set_aspect('equal', adjustable='box')
plt.show()

# 输出文件路径
excel_path
```

```python
import numpy as np
from scipy.integrate import quad

# 等距螺线的参数
a = 0  # 初始半径
b = 0.55 / (2 * np.pi)  # b参数由螺距p=55cm确定

# 定义弧长的积分函数
def arc_length(theta):
    r = a + b * theta
    dr_dtheta = b
    return np.sqrt(dr_dtheta**2 + r**2)

# 积分计算从theta1到theta2的弧长
theta1 = 0  # 起始角度
theta2 = 2 * np.pi * 16  # 16圈

# 使用SciPy的quad进行数值积分
length, error = quad(arc_length, theta1, theta2)

# 输出弧长
print(f"弧长为: {length} 米")
```

```python
import numpy as np
from scipy.integrate import quad
from scipy.optimize import fsolve

# 等距螺线的参数
a = 0  # 初始半径
b = 55 / (2 * np.pi)  # b参数由螺距p=55cm确定

# 定义弧长的积分函数
def arc_length(theta):
    r = a + b * theta
    dr_dtheta = b
    return np.sqrt(dr_dtheta**2 + r**2)

# 定义目标函数，求解下限
def objective(theta1):
    theta2 = 2 * np.pi * 16
    length, _ = quad(arc_length, theta1, theta2)
    return length - 1  # 我们希望弧长为1

# 使用fsolve寻找下限theta1
theta1_initial_guess = 0
theta1_solution = fsolve(objective, theta1_initial_guess)

print(f"积分下限为: {theta1_solution[0]} 弧度")
```

```python
import numpy as np

def theta_xiaxian(beta):
    # 等距螺线的参数
    a = 0  # 初始半径
    b = 0.55 / (2 * np.pi)  # b参数由螺距p=55cm确定

    # 定义弧长的积分函数
    def arc_length(theta):
        r = a + b * theta
        dr_dtheta = b
        return np.sqrt(dr_dtheta**2 + r**2)
    
    # 定义目标函数，求解下限alpa
    def objective(alpa):
        length, _ = quad(arc_length, alpa, beta)
        return length - 1  # 我们希望弧长为1

    # 使用fsolve寻找下限alpa
    theta1_initial_guess = 0
    theta1_xia = fsolve(objective, theta1_initial_guess)
    return theta1_xia[0]

point = []
i = 0
theta_shang = 2 * np.pi * 16

while True:
    point.append([i, theta_shang])
    theta_shang = theta_xiaxian(theta_shang)
    i += 1
    if theta_shang <= 0:
        break
    

print("Points:", point)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_spiral_with_points(point):
    # 等距螺线的参数
    p = 0.55  # 螺距
    b = p / (2 * np.pi)  # b参数由螺距p确定
    a = 0  # 初始半径

    # 生成螺线的角度范围
    theta = np.linspace(0, 2 * np.pi * 16, 1000)  # 从0到16圈

    # 计算螺线半径
    r = a + b * theta

    # 将极坐标转换为笛卡尔坐标
    x_spiral = r * np.cos(theta)
    y_spiral = r * np.sin(theta)

    # 提取点数据
    x_points = []
    y_points = []
    for i, theta_point in point:
        r_point = a + b * theta_point
        x_point = r_point * np.cos(theta_point)
        y_point = r_point * np.sin(theta_point)
        x_points.append(x_point)
        y_points.append(y_point)

    # 绘制螺线
    plt.figure(figsize=(8, 8))
    plt.plot(x_spiral, y_spiral, label="等距螺线")
    
    # 绘制点
    plt.scatter(x_points, y_points, color='red', marker='o', label='点')

    # 图形设置
    plt.title("等距螺线及其上的点")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.gca().set_aspect('equal', adjustable='box')
    plt.grid(True)
    plt.legend()
    plt.show()

# 示例：定义一个点序列

plot_spiral_with_points(point)
```
