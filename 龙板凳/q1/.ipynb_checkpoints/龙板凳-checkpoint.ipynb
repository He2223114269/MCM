```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams

# 
rcParams['font.sans-serif'] = ['SimHei']  # 
rcParams['axes.unicode_minus'] = False  # 

# 
pitch = 55  # cm
b = pitch / (2 * np.pi)  # b 
a = 0  #  0
v_tangent = 1  #  1 
num_turns = 16  # 

# 
total_time = int(2 * np.pi * num_turns * b / v_tangent)  #  =  / 
time_steps = np.arange(0, total_time + 1, 1)  # 

# 
positions = []

# 
for t in time_steps:
    #  theta
    theta_t = v_tangent * t / b
    
    #  r
    r_t = a + b * theta_t
    
    # 
    x_t = r_t * np.cos(theta_t)
    y_t = r_t * np.sin(theta_t)
    
    #  (x, y)
    positions.append([t, x_t, y_t])

#  DataFrame
df = pd.DataFrame(positions, columns=[' ()', 'X  ()', 'Y  ()'])

#  Excel 
excel_path = "data/spiral_positions.xlsx"
df.to_excel(excel_path, index=False)

# 
x_vals = df['X  ()']
y_vals = df['Y  ()']
plt.figure(figsize=(6, 6))
plt.plot(x_vals, y_vals)
plt.title(f" 1 ")
plt.xlabel("X  ()")
plt.ylabel("Y  ()")

#  x  y 
plt.xlim([min(x_vals-100), max(x_vals+100)])
plt.ylim([min(y_vals-100), max(y_vals+100)])

plt.gca().set_aspect('equal', adjustable='box')
plt.show()

# 
excel_path
```

```python
import numpy as np

def theta_xiaxian(beta):
    # 
    a = 0  # 
    b = 0.55 / (2 * np.pi)  # bp=55cm

    # 
    def arc_length(theta):
        r = a + b * theta
        dr_dtheta = b
        return np.sqrt(dr_dtheta**2 + r**2)
    
    # alpa
    def objective(alpa):
        length, _ = quad(arc_length, alpa, beta)
        return length - 1  # 1

    # fsolvealpa
    theta1_initial_guess = 0
    theta1_xia = fsolve(objective, theta1_initial_guess)
    return theta1_xia[0]

point = []
i = 0
theta_shang = 2 * np.pi * 16

while True:
    point.append([i, theta_shang])
    theta_shang = theta_xiaxian(theta_shang)
    i += 1
    if theta_shang <= 0:
        break
    

print("Points:", point)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_spiral_with_points(point):
    # 
    p = 0.55  # 
    b = p / (2 * np.pi)  # bp
    a = 0  # 

    # 
    theta = np.linspace(0, 2 * np.pi * 16, 1000)  # 016

    # 
    r = a + b * theta

    # 
    x_spiral = r * np.cos(theta)
    y_spiral = r * np.sin(theta)

    # 
    x_points = []
    y_points = []
    for i, theta_point in point:
        r_point = a + b * theta_point
        x_point = r_point * np.cos(theta_point)
        y_point = r_point * np.sin(theta_point)
        x_points.append(x_point)
        y_points.append(y_point)

    # 
    plt.figure(figsize=(8, 8))
    plt.plot(x_spiral, y_spiral, label="")
    
    # 
    plt.scatter(x_points, y_points, color='red', marker='o', label='')

    # 
    plt.title("")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.gca().set_aspect('equal', adjustable='box')
    plt.grid(True)
    plt.legend()
    plt.show()

# 

plot_spiral_with_points(point)
```
