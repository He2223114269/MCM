```python
import numpy as np
import pandas as pd
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# 
l = 1.7  # l 
t = 100
pi = np.pi

# 
def system(t, y):
    r, alpha = y
    # 1 d(alpha)/dt
    dalpha_dt = np.sqrt(1 / (r**2 * (1 + (l**2) / (4 * pi**2 * r**2))**2))
    # 2 d(r)/dt
    dr_dt = l * dalpha_dt / (2 * pi)
    return [dr_dt, dalpha_dt]

# 
r0 = 0.001
alpha0 = 0
y0 = [r0, alpha0]

# 
t_span = [0,t]  #  t=0  t=10
t_eval = np.linspace(t_span[0], t_span[1], t)

# 
sol = solve_ivp(system, t_span, y0, t_eval=t_eval)

# 
r_sol = sol.y[0]
alpha_sol = sol.y[1]
t_sol = sol.t

#  x = r*cos(alpha)  y = r*sin(alpha) 
x = - r_sol * np.cos(alpha_sol)
y = - r_sol * np.sin(alpha_sol)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Trajectory (x, y)', color='blue')
plt.title('Trajectory in Cartesian Coordinates')
plt.xlabel('x')
plt.ylabel('y')
# plt.grid(True)
plt.axis('equal')  # 
plt.show()
```
