```python
import math

def calculate_distance(point1, point2):
    """
    计算二维平面上两个点之间的欧几里得距离。
    
    参数:
    point1 (tuple): 第一个点的坐标 (x1, y1)。
    point2 (tuple): 第二个点的坐标 (x2, y2)。
    
    返回:
    float: 两点之间的距离。
    """
    x1, y1 = point1
    x2, y2 = point2
    distance = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
    return distance

# # 示例使用
# point1 = (1, 2)
# point2 = (4, 6)
# dist = calculate_distance(point1, point2)
# print(f"The distance between {point1} and {point2} is: {dist}")
```

```python
def slope_angle(x1, x2, y1, y2):
    # 计算 y 和 x 的差
    delta_y = y2 - y1
    delta_x = x2 - x1
    
    # 计算倾斜角，结果以弧度为单位
    theta = np.arctan2(delta_y, delta_x)
    
    # 将角度限制在 0 到 pi 之间
    if theta < 0:
        theta += np.pi
    elif theta > np.pi:
        theta -= np.pi
       
    return theta# theta_degrees
```
